```python
import pandas as pd
```

```python
date = pd.read_csv("./data/date.csv", sep=';', encoding='utf-8')
card = pd.read_csv("./data/card.csv", sep=';', encoding='utf-8')
seller = pd.read_csv("./data/seller.csv", sep=';', encoding='utf-8')
card_stats = pd.read_csv("./data/card_stats.csv", sep=';', encoding='utf-8')

so_1 = pd.read_csv("./data/sale_offer.csv", sep=';', encoding='utf-8')
so_2 = pd.read_csv("./data/sale_offer_2.csv", sep=';', encoding='utf-8')
so_3 = pd.read_csv("./data/sale_offer_3.csv", sep=';', encoding='utf-8')
so_4 = pd.read_csv("./data/sale_offer_4.csv", sep=';', encoding='utf-8')
sale_offer = pd.concat([so_1, so_2, so_3, so_4]).reset_index(drop=True)
```

```python
date = date.rename(columns={"date_ID": "id", "day_of_week": "weekday"})
date
```

```python
card = card.rename(columns={"card_ID": "id", "card_name": "name", "expansion_name": "expansion"})
card
```

```python
seller = seller.rename(columns={"seller_ID": "id", "seller_name": "name"})
seller
```

```python
card_stats = card_stats.rename(columns={"card_ID": "card_id", "date_ID": "date_id",
                                        "30_avg_price": "monthly_avg", "7_avg_price": "weekly_avg", "1_avg_price": "daily_avg"})
card_stats = pd.concat([pd.Series(name="id", data=card_stats.index+1), card_stats.drop("id", axis=1)], axis=1)
card_stats
```

```python
card_stats["date_id"] = pd.Series(map(lambda x: int(x), card_stats["date_id"]))
card_stats
```

```python
sale_offer = sale_offer.rename(columns={"seller_ID": "seller_id", "card_ID": "card_id",
                                        "language": "card_language", "date_ID": "date_id"})
sale_offer = pd.concat([pd.Series(name="id", data=sale_offer.index+1), sale_offer], axis=1)
sale_offer
```

```python
date.to_csv("./.pickles/date.csv", sep=';', index=False, encoding='utf-8')
card.to_csv("./.pickles/card.csv", sep=';', index=False, encoding='utf-8')
seller.to_csv("./.pickles/seller.csv", sep=';', index=False, encoding='utf-8')
card_stats.to_csv("./.pickles/card_stats.csv", sep=';', index=False, encoding='utf-8')
sale_offer.to_csv("./.pickles/sale_offer.csv", sep=';', index=False, encoding='utf-8')
```
