```python
import pandas as pd
```

```python
date = pd.read_csv("./data/date.csv", compression='gzip', sep=';', encoding='utf-8')
card = pd.read_csv("./data/card.csv", compression='gzip', sep=';', encoding='utf-8')
seller = pd.read_csv("./data/seller.csv", compression='gzip', sep=';', encoding='utf-8')
card_stats = pd.read_csv("./data/card_stats.csv", compression='gzip', sep=';', encoding='utf-8')
sale_offer = pd.read_csv("./data/sale_offer.csv", compression='gzip', sep=';', encoding='utf-8')
```

```python
new_offers = pd.read_csv("./data/sale_offer.csv", compression='gzip', sep=';', encoding='utf-8')
new_offers
```

```python
old = pd.read_csv("./.data/sale_offer.csv", compression='gzip', sep=';', encoding="utf-8")
old
```

```python
added_card_ids = list(new_offers['card_id'].unique())
tb_dropped = old[old['card_id'].isin(added_card_ids)
                    & (old['date_id'] == 39)].index
old = old.drop(tb_dropped)
old
```

```python
# Merge new offers with previous ones and refresh the index and id
old = pd.concat([old, new_offers]).reset_index(drop=True)
old.index += 1
old["id"] = old.index
old
```

```python
sale_offer = old
sale_offer
```
